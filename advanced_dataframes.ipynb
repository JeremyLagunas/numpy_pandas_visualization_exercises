```python
## 5. Join the employees and titles DataFrames together.

url = get_db_url('employees')
url
sql_emp = 'SELECT * FROM employees'
```

```python
employees = pd.read_sql(sql_emp, url)
employees
```

```python
sql_titles = 'SELECT * FROM titles'
titles = pd.read_sql(sql_titles, url)
titles
```

```python
merged_table = employees.merge(titles, how = 'inner', left_on = 'emp_no', right_on = 'emp_no', )
merged_table
```

```python
## 6. For each title, find the hire date of the employee that was hired most recently with that title.

merged_table.groupby('title').hire_date.max().reset_index()
```
