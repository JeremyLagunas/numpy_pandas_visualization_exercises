```python
## 1. Use matplotlib to plot the following equation: y = x^2 - x + 2
##  Add an anotation for the point 0, 0, the origin.
x = list(range(-50, 51))
y = [(n ** 2) - n + 2 for n in x]
plt.plot(x, y)
plt.annotate('Origin', xy = (0, 0), xytext = (-5, 1000), arrowprops = {'facecolor': 'k'})
```

```python
## 3. Combine the figures you created in the last step into one large figure with 4 subplots.
n_rows = 2
n_cols = 2
```

```python
x = np.array(range(0, 800))
y = np.sqrt(x)
plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y)
plt.title('y = sqrt(x)')

x = np.array(range(-200, 200))
y = x * x * x
plt.subplot(n_rows, n_cols, 2)
plt.plot(x, y)
plt.title('y = x^3')

x = np.array(range(0, 100))
y = 2 ** x
plt.subplot(n_rows, n_cols, 3)
plt.plot(x, y)
plt.title('y = 2^x')

x = np.array(range(0, 50))
y = 1 / (x + 1)
plt.subplot(n_rows, n_cols, 4)
plt.plot(x, y)
plt.title('y = 1 / (x + 1)')
```

```python
## 4. Combine the figures you created in the last step into one figure where each of the 4 equations has a different color for the points. Be sure to include a legend and an appropriate title for the figure.
n_rows = 2
n_cols = 2

plt.figure(figsize = (10, 10))

x = np.array(range(0, 800))
y = np.sqrt(x)
plt.subplot(n_rows, n_cols, 1)
plt.plot(x, y, c = 'green')
plt.title('y = sqrt(x)')

x = np.array(range(-200, 200))
y = x * x * x
plt.subplot(n_rows, n_cols, 2)
plt.plot(x, y, c = 'blue')
plt.title('y = x^3')

x = np.array(range(0, 100))
y = 2 ** x
plt.subplot(n_rows, n_cols, 3)
plt.plot(x, y, c = 'orange')
plt.title('y = 2^x')

x = np.array(range(0, 50))
y = 1 / (x + 1)
plt.subplot(n_rows, n_cols, 4)
plt.plot(x, y, c = 'black')
plt.title('y = 1 / (x + 1)')

plt.suptitle('four graphs')
```
