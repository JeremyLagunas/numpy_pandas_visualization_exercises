```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.special
```

```python
x = np.array(range(1, 19))

plt.figure(figsize = (10, 6))

plt.ylim(-1, 40)
plt.xlim(1, 10)

plt.plot(x, (0*x + 1), label = 'O(1)')
plt.plot(x, np.log(x), label = 'Ologn')
plt.plot(x, x, label = 'O(n)')
plt.plot(x, (x * (np.log(x))), label = 'O(n log n)')
plt.plot(x, (x*x), label = 'O($n^2$)')
plt.plot(x, 2**x, label = 'O($2^n$)')
plt.plot(x, scipy.special.factorial(x), label = 'O(n!)')
plt.plot(x, x**x, label = 'O($n^n$)')

plt.text(1.5, 20, 'Jeremy', size = 'xx-large', weight = 'extra bold')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.title('THE BIG O')
```
